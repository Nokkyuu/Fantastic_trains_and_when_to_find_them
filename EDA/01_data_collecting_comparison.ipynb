# Statistical Tests

```python
from sklearn import preprocessing

x1 = df_val[["arrival_delay_m"]] #returns a numpy array
x2 = df_val67[["arrival_delay_m"]]
min_max_scaler = preprocessing.MinMaxScaler()
x1_scaled = min_max_scaler.fit_transform(x1)
x2_scaled = min_max_scaler.fit_transform(x2)
```

```python
from scipy.stats import mannwhitneyu
stat, p_value = mannwhitneyu(x1_scaled, x2_scaled)
print(f'Statistic: {stat}, p-value: {p_value}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as stats

# Example data (replace these with your actual datasets)
x1_scaled = np.random.exponential(scale=1.0, size=1000)
x2_scaled = np.random.exponential(scale=1.1, size=1000)

# Plotting
fig, axes = plt.subplots(2, 2, figsize=(14, 12))

# Histogram with KDE
sns.histplot(x1_scaled, kde=True, color='blue', label='all', stat='density', bins=30, ax=axes[0, 0])
sns.histplot(x2_scaled, kde=True, color='red', label='wo67', stat='density', bins=30, ax=axes[0, 0])
axes[0, 0].set_title('Histogram with KDE')
axes[0, 0].set_xlabel('Value')
axes[0, 0].set_ylabel('Density')
axes[0, 0].legend()

# Kernel Density Plot
sns.kdeplot(x1_scaled, color='blue', label='all', fill=True, ax=axes[0, 1])
sns.kdeplot(x2_scaled, color='red', label='wo67', fill=True, ax=axes[0, 1])
axes[0, 1].set_title('Kernel Density Plot')
axes[0, 1].set_xlabel('Value')
axes[0, 1].set_ylabel('Density')
axes[0, 1].legend()

# Q-Q plot for x1_scaled
stats.probplot(x1_scaled, dist="expon", plot=axes[1, 0])
axes[1, 0].set_title('Q-Q Plot for all')

# Q-Q plot for x2_scaled
stats.probplot(x2_scaled, dist="expon", plot=axes[1, 1])
axes[1, 1].set_title('Q-Q Plot for wo67')

plt.tight_layout()
plt.show()
```
