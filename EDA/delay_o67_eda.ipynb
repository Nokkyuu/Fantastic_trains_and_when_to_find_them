```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
sns.histplot(data=df, x="arrival_delay_m", bins=2000)
plt.title("Histogram Arrival delay")
plt.xlabel("Delay")
plt.ylabel("Frequency")
plt.ylim(0, 2500)
plt.xlim(-20, 40)
plt.show()
```

```python
df_val = df[df["arrival_delay_m"] > 0]

sns.histplot(data=df_val, x="arrival_delay_m", bins=1500)
plt.title("Histogram Arrival delay")
plt.xlabel("Delay")
plt.ylabel("Frequency")
plt.ylim(0, 1000)
plt.xlim(0, 40)
plt.show()
```

```python
df.head()
```

```python
df_stations = df_val.groupby("name").sum("arrival_delay_m")
```

```python
df_stations.head()
```

```python
#Plotting the locations for the picked houses like we did for the whole dataset
fig = px.scatter_mapbox(df_stations, lat='lat', lon='long', hover_name='eva_nr',
                        color='zip', 
                        size_max=20, zoom=10,
                        #size="arrival_delay_m")
)

# Update layout and setting the map
fig.update_layout(mapbox_style='open-street-map',
                  width = 1000,
                  height = 800)

# Show the plot
fig.show()
```
